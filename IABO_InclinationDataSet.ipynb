Se abre un documento de excel y se segmentan y se calcula la media de las señales de la clase IZQUIERDA.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xlrd

wb=xlrd.open_workbook("/content/IABO_Inclinations/Izquierda.xlsx")
sheet = wb.sheet_by_index(0)
#sheet.cell_value(0, 0)
ax=sheet.col_values(0)
ay=sheet.col_values(1)
az=sheet.col_values(2)
gx=sheet.col_values(3)
gy=sheet.col_values(4)
gz=sheet.col_values(5)

N=20.0                      # Número de muestras
nV=np.array(np.shape(ax))   # Número de muestras de la señal
segment = np.floor(nV/N)    # Número de segmetos de la señal
print("Segmentos : ",segment)

m_ventana_axIzq=[]
m_ventana_ayIzq=[]
m_ventana_azIzq=[]
m_ventana_gxIzq=[]
m_ventana_gyIzq=[]
m_ventana_gzIzq=[]

for i in range (0,np.int(segment)):
  ventana_ax=ax[int(i*N):int((i+1)*N)]
  ventana_ay=ay[int(i*N):int((i+1)*N)]
  ventana_az=az[int(i*N):int((i+1)*N)]
  ventana_gx=gx[int(i*N):int((i+1)*N)]
  ventana_gy=gy[int(i*N):int((i+1)*N)]
  ventana_gz=gz[int(i*N):int((i+1)*N)]
  m_ventana_axIzq.append(np.mean(ventana_ax))
  m_ventana_ayIzq.append(np.mean(ventana_ay))
  m_ventana_azIzq.append(np.mean(ventana_az))
  m_ventana_gxIzq.append(np.mean(ventana_gx))
  m_ventana_gyIzq.append(np.mean(ventana_gy))
  m_ventana_gzIzq.append(np.mean(ventana_gz))

plt.subplot(311)
plt.plot(m_ventana_axIzq)
plt.subplot(312)
plt.plot(m_ventana_ayIzq)
plt.subplot(313)
plt.plot(m_ventana_azIzq)
plt.figure()
plt.subplot(311)
plt.plot(m_ventana_gxIzq)
plt.subplot(312)
plt.plot(m_ventana_gyIzq)
plt.subplot(313)
plt.plot(m_ventana_gzIzq)
```

Se unen cada una de las caractiristicas tomadas para obtener la tabla de caracteristicas.

```python
CarAx=m_ventana_axIzq+m_ventana_axDer+m_ventana_axAde+m_ventana_axAtr+m_ventana_axEst
CarAy=m_ventana_ayIzq+m_ventana_ayDer+m_ventana_ayAde+m_ventana_ayAtr+m_ventana_ayEst
CarAz=m_ventana_azIzq+m_ventana_azDer+m_ventana_azAde+m_ventana_azAtr+m_ventana_azEst
CarGx=m_ventana_gxIzq+m_ventana_gxDer+m_ventana_gxAde+m_ventana_gxAtr+m_ventana_gxEst
CarGy=m_ventana_gyIzq+m_ventana_gyDer+m_ventana_gyAde+m_ventana_gyAtr+m_ventana_gyEst
CarGz=m_ventana_gzIzq+m_ventana_gzDer+m_ventana_gzAde+m_ventana_gzAtr+m_ventana_gzEst

caracteristicas=[CarAx,CarAy,CarAz,CarGx,CarGy,CarGz]
caracteristicas=np.array(np.transpose(caracteristicas))
#print(caracteristicas)
```

Se grafica la media de los datos tomados en el giroscopio en X contra los datos tomados con el giroscopio en Y.

```python
plt.scatter(caracteristicas[0:50,0],caracteristicas[0:50,1],alpha=0.3,label='Izquierda',c='red')
plt.scatter(caracteristicas[50:100,0],caracteristicas[50:100,1],alpha=0.3,label='Derecha',c='blue')
plt.scatter(caracteristicas[100:150,0],caracteristicas[100:150,1],alpha=0.3,label='Adelante',c='purple')
plt.scatter(caracteristicas[150:200,0],caracteristicas[150:200,1],alpha=0.3,label='Atras',c='green')
plt.scatter(caracteristicas[200:250,0],caracteristicas[200:250,1],alpha=0.3,label='Estatico',c='yellow')
plt.xlabel("Media_aX")
plt.ylabel("Media_aY")
plt.legend()
```

Se crea un vector de clases para facil identificación en el clasificador.

```python
clases=[]

for n in range(50):
  clases.append(0)

for n in range(50):
  clases.append(1)

for n in range(50):
  clases.append(2)

for n in range(50):
  clases.append(3)

for n in range(50):
  clases.append(4)

clases=np.transpose(clases)

print(clases)
```

Se unen los datos para la posterior creación del excel del DataSet.

```python
# Creación del DataFrame
import pandas as pd

datos={'Media_aX': CarAx,
       'Media_aY': CarAy,
       'Media_aZ': CarAz,
       'Media_gX': CarGx,
       'Media_gY': CarGy,
       'Media_gZ': CarGz,
       'Inclinación':clases}

datosDF=pd.DataFrame(datos)
print(datosDF)
```
